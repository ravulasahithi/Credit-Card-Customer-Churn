# Data Preprocessing & Feature Engineering

Data preprocessing is one of the most important stages in any analytics project.

In this section, we will:

- Verify data quality
- Create business-friendly features
- Transform variables for easier analysis
- Prepare the dataset for visualization and machine learning

```python
data = df.copy()

print("Working copy created successfully.")
```

```python
print(f"Rows    : {data.shape[0]}")
print(f"Columns : {data.shape[1]}")
```

```python
missing = pd.DataFrame({
    "Missing Values": data.isnull().sum(),
    "Percentage": round((data.isnull().sum()/len(data))*100,2)
})

missing.sort_values("Missing Values",ascending=False)
```

Observation

There are no missing values in the dataset.

The dataset is already clean and suitable for further preprocessing.

```python
duplicates = data.duplicated().sum()

print("Duplicate Rows :",duplicates)
```

```python
if duplicates>0:
    data.drop_duplicates(inplace=True)

print(data.shape)
```

```python
data["Churn"] = data["Attrition_Flag"].map({
    "Existing Customer":0,
    "Attrited Customer":1
})
data[["Attrition_Flag","Churn"]].head(5)
```

```python
bins = [20,30,40,50,60,100]

labels = [
    "Under 30",
    "30-40",
    "41-50",
    "51-60",
    "60+"
]

data["Age_Group"] = pd.cut(
    data["Customer_Age"],
    bins=bins,
    labels=labels
)
data["Age_Group"].value_counts()
```

```python
data["Credit_Category"] = pd.qcut(
    data["Credit_Limit"],
    q=4,
    labels=[
        "Low",
        "Medium",
        "High",
        "Premium"
    ]
)
data["Credit_Category"].value_counts()
```

```python
data["Transaction_Level"] = pd.qcut(
    data["Total_Trans_Ct"],
    q=3,
    labels=[
        "Low",
        "Medium",
        "High"
    ]
)
data["Transaction_Level"].value_counts()
```

```python
data["Transaction_Amount_Level"] = pd.qcut(
    data["Total_Trans_Amt"],
    q=4,
    labels=[
        "Low",
        "Medium",
        "High",
        "Very High"
    ]
)
data["Transaction_Amount_Level"].value_counts()
```

```python
bins=[0,24,36,48,60]

labels=[
    "New",
    "Growing",
    "Established",
    "Loyal"
]

data["Customer_Tenure"] = pd.cut(
    data["Months_on_book"],
    bins=bins,
    labels=labels
)
data["Customer_Tenure"].value_counts()
```

```python
def utilization_band(x):

    if x==0:
        return "No Utilization"

    elif x<0.30:
        return "Low (0-30%)"

    elif x<0.70:
        return "Medium (30-70%)"

    else:
        return "High (70%+)"
```

```python
data["Utilization_Band"] = data["Avg_Utilization_Ratio"].apply(utilization_band)
```

```python
data["Utilization_Band"].value_counts()
```

```python
def customer_activity(x):

    if x<=30:
        return "Low"

    elif x<=70:
        return "Medium"

    else:
        return "High"
```

```python
data["Customer_Activity"] = data["Total_Trans_Ct"].apply(customer_activity)
```

```python
data["Customer_Activity"].value_counts()
```

```python
data["Income_Category"] = data["Income_Category"].replace(
    "Unknown",
    "Not Disclosed"
)
```

```python
data["Income_Category"].value_counts()
```

```python
data["Education_Level"] = data["Education_Level"].replace(
    "Unknown",
    "Not Disclosed"
)
```

```python
data["Education_Level"].value_counts()
```

```python
data["Marital_Status"] = data["Marital_Status"].replace(
    "Unknown",
    "Not Disclosed"
)
```

```python
data["Marital_Status"].value_counts()
```

```python
data.info()
```

```python
data.to_csv(
    "Customer_Churn_Cleaned.csv",
    index=False
)

print("Clean dataset exported successfully.")
```

```python
churn_counts = data["Attrition_Flag"].value_counts().reset_index()
churn_counts.columns = ["Customer Status", "Count"]

fig = px.pie(
    churn_counts,
    names="Customer Status",
    values="Count",
    hole=0.55,
    title="Customer Churn Distribution"
)

fig.show()
```

### Recommendation

Focus retention campaigns on high-risk customers before they discontinue their services.

```python
gender = data["Gender"].value_counts().reset_index()
gender.columns = ["Gender", "Customers"]

fig = px.bar(
    gender,
    x="Gender",
    y="Customers",
    text="Customers",
    title="Customer Distribution by Gender"
)

fig.show()
```

# Business Analysis

In this section, we answer key business questions using data.

The objective is to identify:
- High-risk customer segments
- Revenue opportunities
- Churn drivers
- Customer retention strategies

The findings can help Harbor Trust Bank reduce churn and improve customer satisfaction.

```python
#Which Age Group has the highest churn?
age_analysis =(
    data.groupby("Age_Group")
        .agg(
            Total_Customers=("CLIENTNUM","count"),
            Churned_Customers=("Churn","sum")
        )
        .reset_index()
)

age_analysis["Churn Rate (%)"] = round(
    (age_analysis["Churned_Customers"] /
     age_analysis["Total_Customers"])*100,2
)

age_analysis
```

```python
fig = px.bar(
    age_analysis,
    x="Age_Group",
    y="Churn Rate (%)",
    color="Churn Rate (%)",
    text="Churn Rate (%)",
    title="Business Question 1: Which Age Group Churns the Most?"
)

fig.show()
```

### Finding

The age group with the highest churn rate should be prioritized for retention campaigns.

### Recommendation

Offer personalized offers and loyalty benefits to customers in this age segment.

```python
#Which Card Type has the highest churn?
card_analysis = (
    data.groupby("Card_Category")
        .agg(
            Customers=("CLIENTNUM","count"),
            Churned=("Churn","sum")
        )
        .reset_index()
)

card_analysis["Churn Rate (%)"] = round(
    card_analysis["Churned"]/
    card_analysis["Customers"]*100,2
)

card_analysis
```

```python
fig = px.bar(
    card_analysis,
    x="Card_Category",
    y="Churn Rate (%)",
    color="Card_Category",
    text="Churn Rate (%)",
    title="Business Question 2: Card Category vs Churn"
)

fig.show()
```

```python
#Do inactive customers churn more?
inactive_analysis = (
    data.groupby("Months_Inactive_12_mon")
        .agg(
            Customers=("CLIENTNUM","count"),
            Churned=("Churn","sum")
        )
        .reset_index()
)

inactive_analysis["Churn Rate (%)"] = round(
    inactive_analysis["Churned"]/
    inactive_analysis["Customers"]*100,
    2
)

inactive_analysis
```

```python
fig = px.line(
    inactive_analysis,
    x="Months_Inactive_12_mon",
    y="Churn Rate (%)",
    markers=True,
    title="Business Question 4: Customer Inactivity"
)

fig.show()
```

```python
#Are customers with fewer products more likely to churn?
relationship_analysis = (
    data.groupby("Total_Relationship_Count")
        .agg(
            Customers=("CLIENTNUM","count"),
            Churned=("Churn","sum")
        )
        .reset_index()
)

relationship_analysis["Churn Rate (%)"] = round(
    relationship_analysis["Churned"]/
    relationship_analysis["Customers"]*100,
    2
)

relationship_analysis
```

```python
fig = px.line(
    relationship_analysis,
    x="Total_Relationship_Count",
    y="Churn Rate (%)",
    markers=True,
    title="Business Question 5: Products Owned vs Churn"
)

fig.show()
```

```python
#Does lower transaction activity indicate churn?
transaction_analysis = (
    data.groupby("Transaction_Level")
        .agg(
            Customers=("CLIENTNUM","count"),
            Churned=("Churn","sum")
        )
        .reset_index()
)

transaction_analysis["Churn Rate (%)"] = round(
    transaction_analysis["Churned"]/
    transaction_analysis["Customers"]*100,
    2
)

transaction_analysis
```

```python
fig = px.bar(
    transaction_analysis,
    x="Transaction_Level",
    y="Churn Rate (%)",
    color="Transaction_Level",
    title="Business Question 6: Transaction Activity"
)

fig.show()
```

```python
#Which customer segment is the highest risk?
risk = (
    data.groupby(
        [
            "Age_Group",
            "Card_Category"
        ]
    )
    .agg(
        Customers=("CLIENTNUM","count"),
        Churned=("Churn","sum")
    )
    .reset_index()
)

risk["Churn Rate (%)"] = round(
    risk["Churned"]/
    risk["Customers"]*100,
    2
)

risk.sort_values(
    "Churn Rate (%)",
    ascending=False
).head(10)
```

```python
fig = px.treemap(
    risk,
    path=["Age_Group","Card_Category"],
    values="Customers",
    color="Churn Rate (%)",
    title="Highest Risk Customer Segments"
)

fig.show()
```
